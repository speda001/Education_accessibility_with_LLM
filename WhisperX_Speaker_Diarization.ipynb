```python
def gather_words_by_speaker(data, speakers):
    # Initialize a dictionary to store lists of words by speaker
    speaker_dict = {speaker: [] for speaker in speakers}

    # Iterate over each segment in the data
    for segment in data['segments']:
        # print("Segment is:", segment)
        speaker = segment.get('speaker', None)  # Safely get the 'speaker' key
        if not speaker:
            print("Warning: Missing 'speaker' key in segment. Skipping segment.")
            continue

        # print("Speaker is", speaker)
        if speaker in speaker_dict:
            # Safely get the 'words' key, defaulting to an empty list if not found
            words = segment.get('words', [])
            speaker_dict[speaker].extend([word['word'] for word in words])

    return speaker_dict

# Example usage:
data = result
# Define the unique speakers
speakers = speakers_list
transcript ={}
words_by_speaker = gather_words_by_speaker(data, speakers)
print(type(words_by_speaker))
transcript = {}

for speaker, words in words_by_speaker.items():
    concatenated_words = " ".join(words)  # Correct way to form a sentence from words
    print(f"{speaker}: {concatenated_words}")
    transcript[speaker] = concatenated_words

print(transcript)  # Print the complete transcript dictionary

from google.colab import files




with open('summary_responses_5.txt', 'w') as file:
    for speaker, text in transcript.items():
      file.write(f"{speaker}:\n{text}\n\n")

# Download the file
files.download('summary_responses_5.txt')
```
